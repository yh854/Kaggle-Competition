```python
df_train_new = df_train[['State', 'License Type', 'Violation', 'Precinct', 'County', 'Issuing Agency', 'Violation Status']]
df_train_new.head()
```

```python
df_train_new.isnull().sum()
```

```python
df_train_new2 = df_train_new.dropna(axis = 0)
df_train_new2.info()
```

```python
df_train_new2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# one hot encoding
X_train = pd.get_dummies(df_train_new2.iloc[:, :-1])

# convert labels to integers
le = LabelEncoder()
y_train = le.fit_transform(df_train_new2['Violation Status'])

print(y_train)
X_train.head()
```

```python
df_test = pd.read_csv('./test.csv')
df_test_new = df_test[['State', 'License Type', 'Violation', 'Precinct', 'County', 'Issuing Agency']]
df_test_new.isnull().sum()
```

```python
df_test.info()
```

```python
for s in df_test['State'].unique():
    if s not in df_train['State'].unique():
        print(s)
        
print(len(df_train['State'].unique()))
print(len(df_test['State'].unique()))
```

```python
for lt in df_test['License Type'].unique():
    if lt not in df_train['License Type'].unique():
        print(lt)
        
print(len(df_train['License Type'].unique()))
print(len(df_test['License Type'].unique()))
```

```python
# one hot encoding
X_test = pd.get_dummies(df_test_new.iloc[:, :])
X_test.head()
```

```python
# 注意要改loop
X_train['State_GV'] = 0
X_train['State_HI'] = 0
X_train['State_AK'] = 0

X_train['License Type_SOS'] = 0
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_test = clf.predict(X_test)
print(y_test)
```
