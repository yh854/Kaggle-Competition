```python
from sklearn.preprocessing import LabelEncoder

# one hot encoding
X_train = pd.get_dummies(df_train_new2.iloc[:, :-1])

# convert labels to integers
le = LabelEncoder()
y_train = le.fit_transform(df_train_new2['Violation Status'])

print(y_train)
X_train.head()
```

```python
df_test = pd.read_csv('./test.csv')
df_test_new = df_test[['State', 'License Type', 'Violation', 'Precinct', 'County', 'Issuing Agency']]
df_test_new.isnull().sum()
```

```python
for v in df_test['Issuing Agency'].unique():
    if v not in df_train['Issuing Agency'].unique():
        print(v)
        
print(len(df_train['Issuing Agency'].unique()))
print(len(df_test['Issuing Agency'].unique()))
```

```python
# one hot encoding
X_test = pd.get_dummies(df_test_new.iloc[:, :])
X_test.head()
```

```python
count = 0
for s in df_test['State'].unique():
    if s not in df_train['State'].unique():
        X_train['State_'+s] = 0
        count += 1
        if count == 3:
            break
            
count = 0
for s in df_test['License Type'].unique():
    if s not in df_train['License Type'].unique():
        X_train['License Type_'+s] = 0
        count += 1
        if count == 9:
            break

count = 0
for s in df_test['Violation'].unique():
    if s not in df_train['Violation'].unique() and pd.isnull(s) == False:
        X_train['Violation_'+s] = 0
        count += 1
        if count == 4:
            break

count = 0
for s in df_test['Issuing Agency'].unique():
    if s not in df_train['Issuing Agency'].unique() and pd.isnull(s) == False:
        X_train['Issuing Agency_'+s] = 0
        count += 1
        if count == 3:
            break
```
